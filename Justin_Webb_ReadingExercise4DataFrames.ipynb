## Joining two DataFrames

[Explanation](https://drive.google.com/file/d/15xUxzjSaQxsZPXXWimpi3w6QGEp596fC/view?usp=drive_link)

```python
dfAQI = pd.read_csv('aqi.txt')

dfAQI = dfAQI.set_index('color').drop('Green')

dfAQIdays = pd.read_csv('aqidays.txt')
dfAQIdays = dfAQIdays.set_index('color').drop('Maroon')

print(dfAQI)
print()
print(dfAQIdays)
```

```python
dfAQI.index
```

```python
result1 = dfAQI.join(dfAQIdays)
result1
```

```python
dfAQIdays.join(dfAQI)
```

```python
dfAQI.join(dfAQIdays, how='inner')
```

```python
dfAQI.join(dfAQIdays, how='outer')
```

- join()'s how parameter controls which rows will be in the result
- Assume: dfLeft.join(dfRight)
- Following table summarizes the resulting rows

 - join type	Included rows
 - left	Include all rows from dfLeft, and matching rows from dfRight (default)
 - right	Include all rows from dfRight, and matching rows from dfLeft
 - outer	Include the union of all rows in dfLeft and dfRight
 - inner	Include rows where index values are in both dfLeft and dfRight
